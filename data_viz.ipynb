```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Bar Plot

```python
# plt.figure(figsize=(12,6))
plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.bar(df['Sales_Date'], df['Sales_Amount'], width=0.5, label= 'Sales')
plt.bar(df['Sales_Date'], df['Profit_Amount'], width=0.5, color="red", label= 'Profit')

plt.legend()

plt.show()
```

```python

positions = np.arange(len(df))
width = 0.5


plt.bar(positions - width/2, df['Sales_Amount'], width=width, label='Sales')
plt.bar(positions + width/2, df['Profit_Amount'], width=width, color='red', label='Profit')


plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.xticks(positions, df['Sales_Date'], rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```
