```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Bar Plot

```python
# plt.figure(figsize=(12,6))
plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.barh(df['Sales_Date'], df['Sales_Amount'],label= 'Sales')
plt.barh(df['Sales_Date'], df['Profit_Amount'],color="red", label= 'Profit')

plt.legend()

plt.show()
```

```python
#using vectorized operations
positions = np.arange(len(df))
width = 0.4

plt.bar(positions - width/2, df['Sales_Amount'], width=width, label='Sales')
plt.bar(positions + width/2, df['Profit_Amount'], width=width, color='red', label='Profit')

plt.xticks(positions, df['Sales_Date'], rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#using offset positions
width = 0.4
pos1 = np.arange(len(df))
pos2 = [j+width for j in pos1]


plt.bar(pos1, df['Sales_Amount'], width=width, label='Sales')
plt.bar(pos2, df['Profit_Amount'], width=width, color='red', label='Profit')


plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.xticks(pos1, df['Sales_Date'], rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```
