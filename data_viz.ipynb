```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\EDA\Matplotlib-Data-VIz\sales_data.csv")
```

### 1. Bar Plot

```python
# plt.figure(figsize=(12,6))
plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.barh(df['Sales_Date'], df['Sales_Amount'],label= 'Sales')
plt.barh(df['Sales_Date'], df['Profit_Amount'],color="red", label= 'Profit')

plt.legend()

plt.show()
```

```python
#using vectorized operations
positions = np.arange(len(df))
width = 0.4

plt.bar(positions - width/2, df['Sales_Amount'], width=width, label='Sales')
plt.bar(positions + width/2, df['Profit_Amount'], width=width, color='red', label='Profit')

plt.xticks(positions, df['Sales_Date'], rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#using offset positions
width = 0.4
pos1 = np.arange(len(df))
pos2 = [j+width for j in pos1]


plt.bar(pos1, df['Sales_Amount'], width=width, label='Sales')
plt.bar(pos2, df['Profit_Amount'], width=width, color='red', label='Profit')


plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.xticks(pos1, df['Sales_Date'], rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```

### 4. Pie Plot

```python
df.head()
```

```python
#product sales
plt.pie(df['Sales_Amount'], labels=df['Product_Name'], explode=[0,0,0,0.1,0], autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
#category sales
grouped_category = df.groupby('Category')['Sales_Amount'].sum()
plt.pie(grouped_category, labels=df['Category'].unique(), shadow=True, radius=1.5)    #or labels = grouped_category.index
plt.pie([1], radius= 1, colors='w')    # [1] is list with only one element
plt.legend(loc='center')
plt.show()
```

### 5. Stem Plot

```python
df.head()
```

```python
#Sales By Date
plt.stem(df.Sales_Date, df.Sales_Amount,
         bottom=400,    # bottom default 0
         linefmt=':',   # : is dotted line for stem
         markerfmt='ro' )   #ro is red circle for marker
plt.show()
```

### 6. Box Plot

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
numeric_df.boxplot(
    vert=True,               # Vertical boxplot
    patch_artist=True,       # Filled boxes
    showmeans=True,          # Shows mean
    boxprops=dict(facecolor='lightblue', color='blue'), # Box color
    whiskerprops=dict(color='green'), # Whisker color
    capprops=dict(color='red'), # Cap color
    flierprops=dict(marker='o', color='red', markersize=10), # Outlier properties
    medianprops=dict(color='purple') # Median line color
)
plt.title('Boxplot for Numeric Columns')
plt.xlabel('Numeric Columns')
plt.ylabel('Values')
plt.show()
```

### 7. Stack Plot

```python
x = [1,2,3,4]
area1 = [4,7,3,1]
area2 = [7,2,2,1]
area3 = [5,7,9,1]
label = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=label)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

### 8. Step Plot

```python
x = [1,2,3,4]
y = [4,7,3,1]
plt.step(x,y, marker='o', mfc='r', ms=10, label='y')    #mfc - color of marker & ms - size of marker
plt.legend(loc=3)           #2    1
plt.show()                  #3    4    location for labels in chart
```

### 9. Linear Plot with Fill Between 

```python
x = np.array([1,2,3,4])
y = np.array([4,7,3,1])    #where conditions do not work with list so array conversion
plt.plot(x,y, color='r')
plt.fill_between(x,y, alpha=0.5, where= (x>=2) & (x<=4))   # alpha for opacity
plt.show()
```

### 10. Using Subplot and Savefig

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.title('Stack Plot')
x = [1,2,3,4]
area1 = [4,7,3,1]
area2 = [7,2,2,1]
area3 = [5,7,9,1]
label = ['area1', 'area2', 'area3']
plt.stackplot(x,area1,area2,area3,labels=label)

plt.subplot(2,2,2)
plt.title('Step Plot')
plt.step(x,y, marker='o', mfc='r', ms=10, label='y')

plt.subplot(2,2,3)
plt.title('Stem Plot')
plt.stem(df.Sales_Date, df.Sales_Amount,
         bottom=400,    # bottom default 0
         linefmt=':',   # : is dotted line for stem
         markerfmt='ro' )   #ro is red circle for marker

plt.subplot(2,2,4)
plt.title('Pie Plot')
plt.pie(df['Sales_Amount'], labels=df['Product_Name'], explode=[0,0,0,0.1,0], autopct='%1.1f%%')

plt.tight_layout()
# plt.savefig('Matplotlib Charts.pdf', dpi=2000, bbox_inches='tight')
plt.show()
```

### 11. Using xticks, xlim and axis functions

```python
x = [1,2,3,4]
y = [4,7,3,1]
plt.plot(x,y)
plt.xticks(x,labels=['Java', 'DBMS', 'OS', 'Python'])

plt.xlabel('Courses')
plt.ylabel('No. of attendees')

plt.ylim(1,10)   #for custom range limit in axis
plt.axis([1,10,1,10])  #to define limit for both x and y axis (min_x,max_x,min_x,max_y)

plt.show()
```
