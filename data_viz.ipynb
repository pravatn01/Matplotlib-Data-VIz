### 1. Bar Plot

```python
# plt.figure(figsize=(12,6))
plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.barh(df['Sales_Date'], df['Sales_Amount'],label= 'Sales')
plt.barh(df['Sales_Date'], df['Profit_Amount'],color="red", label= 'Profit')

plt.legend()

plt.show()
```

```python
#using vectorized operations
positions = np.arange(len(df))
width = 0.4

plt.bar(positions - width/2, df['Sales_Amount'], width=width, label='Sales')
plt.bar(positions + width/2, df['Profit_Amount'], width=width, color='red', label='Profit')

plt.xticks(positions, df['Sales_Date'], rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#using offset positions
width = 0.4
pos1 = np.arange(len(df))
pos2 = [j+width for j in pos1]


plt.bar(pos1, df['Sales_Amount'], width=width, label='Sales')
plt.bar(pos2, df['Profit_Amount'], width=width, color='red', label='Profit')


plt.title("Sales By Date", fontsize=15)
plt.xlabel("Date")
plt.ylabel("Amount")

plt.xticks(pos1, df['Sales_Date'], rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```

### Pie Plot

```python
df.head()
```

```python
#product sales
plt.pie(df['Sales_Amount'], labels=df['Product_Name'], explode=[0,0,0,0.1,0], autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
#category sales
grouped_category = df.groupby('Category')['Sales_Amount'].sum()
plt.pie(grouped_category, labels=df['Category'].unique(), shadow=True, radius=1.5)    #or labels = grouped_category.index
plt.pie([1], radius= 1, colors='w')    # [1] is list with only one element
plt.legend(loc='center')
plt.show()
```
